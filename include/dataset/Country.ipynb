```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.iban.com/country-codes'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', class_='table-bordered')
table
```

```python
table_titles = table.find_all('th')
table_titles
```

```python
clean_table_titles = [title.text.strip() for title in table_titles]
clean_table_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=clean_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.rename(columns={'Alpha-2 code': 'Alpha-2_code', 'Alpha-3 code': 'Alpha-3_code'}, inplace=True)
df
```

```python
df.to_csv(r'./country_codes.csv', index=False)
```
