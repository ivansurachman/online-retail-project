```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://docs.dnb.com/partner/en-US/iso_country_codes'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', class_='TableStyle-dnb_tablestyle_2')
table
```

```python
table_titles = table.find_all('tr')[0]
table_titles
```

```python
clean_table_titles = [' '.join(title.text.replace('\n', ' ').split()) for title in table_titles]
clean_table_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=clean_table_titles)
df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df_cleanCol = df.drop(['WorldBase Country Name', 'WorldBase 3 digit', 'WorldBase 2 char', 'Comments'], axis=1)
df_cleanCol
```

```python
print(f"Shape of df_cleanCol before removing duplicates: {df_cleanCol.shape}")
df_cleanCol_no_duplicates = df_cleanCol.drop_duplicates()
print(f"Shape of df_cleanCol after removing duplicates: {df_cleanCol_no_duplicates.shape}")
display(df_cleanCol_no_duplicates.head())
```

```python
df_cleanCol_no_duplicates = df_cleanCol_no_duplicates.rename(columns={"ISO Country Name":"ISO_Country_Name", "ISO 3 digit":"ISO_3_digit", "ISO 3 char":"ISO_3_char", "ISO 2 char":"ISO_2_char"})
df_cleanCol_no_duplicates
```

```python
df_cleanCol_no_duplicates.to_csv(r'./country_codes.csv', index=False)
```
